## Evaluation

### Technology Trend - 2016

```python
def SplitTechs(val, tech, seperator):
    '''
        Input:
            val - String to be inspcted
            tech - Specific thechnology whose presence in the "val" is evaluated
            separaor - data separator
        Outpt:
            0 or 1 based on the prescence of tech in val
    '''
    try:
        tech_list = val.split(seperator)
        if tech in tech_list:
            return 1
        else:
            return 0
    except AttributeError:
        return 0
```

```python
def SplitTechnologyData(techlist,df_response, year, seperator):
    '''
        Input: 
            techlist : List of technologies whose prescence needs to be evaluated
            df_response: Data frame cotaining the data
            year: year in which the survey is conducted
            separator: sepeator used for seggragating the technologies
        
    '''
    techData=[]
    techData = pd.DataFrame(techData)
    # Apply function for each Languague
    for tech in techlist:
        techData[tech] = np.vectorize(SplitTechs)(df_response, tech, seperator)
        
    valCounts = techData.sum().sort_values(ascending=False)
    x = valCounts.index
    y = valCounts.values

    df_general = pd.DataFrame({'Tech' : x, year : y})
    df_general.plot(x='Tech', y = year, kind='bar', figsize=(20,5))
    plt.ylabel('User Count')
    plt.show()

    return techData, valCounts
```
