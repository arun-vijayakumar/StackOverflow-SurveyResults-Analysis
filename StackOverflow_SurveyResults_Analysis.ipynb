## Trends in Salary

```python
#2016
#print(df_response_2016['occupation'].unique())
#print(df_response_2017['EmploymentStatus'].unique())
#print(df_response_2016['job_discovery'].unique())
#print(df_response_2016['new_job_value'].unique())
#print(df_response_2017['LastNewJob'].unique())
#print(df_response_2018['LastNewJob'].unique())
#salary_midpoint
#CompanySize
#CompanyType
#YearsProgram
#Salary
salary_2016 = df_response_2016['salary_midpoint']

salary_2017 = df_response_2017['Salary']
salary_2017 = salary_2017.apply(lambda x: round(x,-3))

salary_2018_raw = df_response_2018[['SalaryType','ConvertedSalary']]



def UpdateSalary(x):
    if x[0] == 'Weekly':
        val = x[1] * 52
    elif x[0] == 'Monthly':
        val = x[1] * 12
    else:
        val = x[1]
    if val < 100000:
        return val
    else:
        return np.nan

salary_2018_raw['YearlySalary'] = salary_2018_raw.apply(UpdateSalary, axis=1)
salary_2018 = salary_2018_raw['YearlySalary']
#salary_2018 = pd.Series(salary_2018)
salary_2018 = salary_2018.apply(lambda x: round(x,-3))
#salary_2018 = salary_2018[salary_2018['YearlySalary' < 1000000]]
#print(salary_2018)
print('User Count:',salary_2016.count())
print('User Count:',salary_2017.count())
print('User Count:',salary_2018.count())

plt.hist(salary_2016, bins=20)
plt.ylabel('Users')
plt.xlabel('Average Salary - 2016')
plt.show()

plt.hist(salary_2017, bins=20)
plt.ylabel('Users')
plt.xlabel('Average Salary - 2017')
plt.show()

plt.hist(salary_2018, bins=20)
plt.ylabel('Users')
plt.xlabel('Average Salary - 2018')
plt.show()
```
